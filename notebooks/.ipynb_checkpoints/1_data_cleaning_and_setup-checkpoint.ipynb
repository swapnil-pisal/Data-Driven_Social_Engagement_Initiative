# Create the Target Variable

```python
# Find the median value of likes
median_likes = df['likes'].median()
print(f"The median number of likes is: {median_likes}")
```

```python
# Create the binary target variable
df['high_engagement'] = np.where(df['likes'] > median_likes, 1, 0)
```

```python
print(df['high_engagement'].value_counts())
```

```python
# Check the distribution to make sure it's balanced
print(df['high_engagement'].value_counts(normalize=True))
```

# Exploratory Data Analysis

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot 1: Does hashtag count affect engagement?
plt.figure(figsize=(10, 5))
sns.boxplot(x='high_engagement', y='hashtags_count', data=df)
plt.title('Hashtag Count vs Engagement Level')
plt.show()

# Plot 2: Which hours of the day get highest engagement?
plt.figure(figsize=(12, 5))
sns.countplot(x='post_hour', hue='high_engagement', data=df)
plt.title('Engagement Distribution Across Hours of the Day')
plt.show()
```
